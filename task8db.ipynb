```python
import sqlite3
import zipfile
import io
import os

# Database path
db_path = "/content/drive/MyDrive/Copy of eng_subtitles_database.db"
extract_path = "/content/newfolder"

# Ensure extraction directory exists
os.makedirs(extract_path, exist_ok=True)

# Connect to the database
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# Fetch all subtitles from the database
cursor.execute("SELECT name, content FROM zipfiles")
rows = cursor.fetchall()

total_files = len(rows)
print(f"Total subtitle files found: {total_files}")

for file_name, blob_data in rows:
    try:
        # Convert BLOB to file-like object
        blob_stream = io.BytesIO(blob_data)

        # Extract ZIP contents
        with zipfile.ZipFile(blob_stream, 'r') as zip_ref:
            zip_ref.extractall(extract_path)

        print(f"✅ Extracted: {file_name}")

    except zipfile.BadZipFile:
        print(f"❌ Skipped (Invalid ZIP): {file_name}")

# Close the database connection
conn.close()

print(f"✅ All subtitles extracted to: {extract_path}")
```

```python
import sqlite3
import zipfile
import io
import os

# ✅ Step 1: Database Path
db_path = "/content/drive/MyDrive/Copy of eng_subtitles_database.db"

# ✅ Step 2: Extraction Folder (Sab kuch yahan save hoga)
extract_path = "/content/datawewant"

# ✅ Step 3: Connect to Database
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# ✅ Step 4: Fetch All Subtitle Files from Database
cursor.execute("SELECT name, content FROM zipfiles")
rows = cursor.fetchall()

# ✅ Step 5: Extraction Process
total_files = len(rows)
failed_extractions = []
print(f"📌 Total Files in Database: {total_files}")
print(f"🚀 Extraction Started...")

for file_name, blob_data in rows:
    try:
        # ✅ Convert BLOB data to file-like object
        blob_stream = io.BytesIO(blob_data)

        # ✅ Extract ZIP contents
        with zipfile.ZipFile(blob_stream, 'r') as zip_ref:
            zip_ref.extractall(extract_path)  # ✅ Save in /content/drive/datawewant/

        print(f"✅ Extracted: {file_name}")

    except zipfile.BadZipFile:
        print(f"❌ Skipped (Corrupt ZIP): {file_name}")
        failed_extractions.append(file_name)

# ✅ Step 6: Close Database Connection
conn.close()

# ✅ Step 7: Print Summary
print(f"\n🎯 Extraction Completed!")
print(f"✅ Successfully Extracted: {total_files - len(failed_extractions)}")
print(f"❌ Failed ZIPs: {len(failed_extractions)}")
print(f"📂 Extracted Files Saved In: {extract_path}")
```

```python
import os
import zipfile
import hashlib
import shutil

# Define paths
zip_file_path = "/content/drive/MyDrive/datawewant.zip"  # Adjust path as needed
extract_path = "content/extracted_datawewant"
cleaned_path = "content/cleaned_datawewant"

# Step 1: Extract ZIP
print("📂 Extracting ZIP file...")
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("✅ Extraction Complete!")

# Step 2: Collect required file formats
valid_extensions = {".srt", ".vtt", ".ass", ".nfo"}
file_hashes = set()

# Ensure cleaned directory exists
os.makedirs(cleaned_path, exist_ok=True)

# Step 3: Process and remove duplicates
print("🧹 Cleaning duplicate files...")
for root, _, files in os.walk(extract_path):
    for file in files:
        if any(file.lower().endswith(ext) for ext in valid_extensions):
            file_path = os.path.join(root, file)

            # Generate file hash for duplicate detection
            with open(file_path, "rb") as f:
                file_hash = hashlib.md5(f.read()).hexdigest()

            # If hash is unique, copy file
            if file_hash not in file_hashes:
                file_hashes.add(file_hash)
                shutil.copy(file_path, os.path.join(cleaned_path, file))
                print(f"✅ Copied: {file}")
            else:
                print(f"❌ Skipped (Duplicate): {file}")

print("🎉 Cleanup complete! Unique files saved in 'cleaned_datawewant' ✅")
```
